Basically `split_by_folder` splits a large `ItemList` (or any of `ItemList`'s subclass) object created with the main folder, into two `ItemList` objects corresponding to two subfolders.

Behind the scenes `split_by_folder` utilizes `_get_by_folder(name)` to creates a list of idxs for all the items of the large `ItemList` which are corresponding to data files of a subfolder named 'tranining' if set `name='training'`.

Behind the scenes, `split_by_idxs` uses a list of indexes assigned to the name `train_idx` and a list of indexes assigned to the name `valid_idx`, to split a large `ItemList` object into two `ItemList` objects attached to an `ItemLists` object.

How to use `ItemList[idxs]` (or how does `ItemList.__getitem__` work)?

> `ItemList[idxs]`'s `idxs` can be a integer or a list of them. When `idxs` is a single integer, it returns a single item of the `ImageList`. When `idxs` is a list of integers, then it returns a new `ItemList` containing `items` selected by those `idxs`.

In fact, `split_by_list` uses `_split` (behind the scene, it is `ItemLists.__init__`) to attach two `ItemList` objects or any of `ItemList`'s subclass objects to an `ItemLists` object.

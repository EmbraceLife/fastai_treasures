# Initial training

```python
fv.np.random.seed(42) # make sure the same validation set
data = fv.ImageDataBunch.from_folder(path_train,
                                     test='../test', # same level with path_train
                                     ds_tfms=fv.get_transforms(),
                                     valid_pct=0.2,
                                     size=299, # why?
                                     bs=16,
                                     num_workers=2)
data.normalize(fv.imagenet_stats)

data.valid_ds.y[:5]
```

```python
learn = fv.create_cnn(data, fv.models.resnet50, metrics=fv.error_rate, model_dir = base_dir)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
# how to choose its learning rate? 
learn.fit_one_cycle(6, max_lr=slice(1e-3,1e-2)) 
```

```python
learn.save("SeedRes152_default_6")
```

```python
learn.fit_one_cycle(10) # default 1e-3
```

```python
learn.save("SeedRes152_default_16")
```

# check lr for fine tuning

```python
learn.fit_one_cycle(4,max_lr=slice(6e-7,1e-4))
```

# Train with different sizes of images

```python
fv.np.random.seed(42)
data = fv.ImageDataBunch.from_folder(path,test='../test', ds_tfms=fv.get_transforms(),
                                  valid_pct=0.2,size=350,bs=16,num_workers=2)
data.normalize(fv.imagenet_stats)

data.valid_ds.y[:5]
```

```python
learn = fv.create_cnn(data, fv.models.resnet152, metrics=fv.error_rate, model_dir = base_dir)
learn.load("SeedRes152_default_16")
```

```python
learn.unfreeze()
learn.fit_one_cycle(10,max_lr=slice(1e-6,1e-4))
```

```python
learn.save("SeedRes152_unfreeze_sizeD_26")
```

```python
learn = fv.create_cnn(data, fv.models.resnet152, metrics=fv.error_rate, model_dir = base_dir)
learn.load("SeedRes152_default_16")
learn.unfreeze()
learn.fit_one_cycle(4,max_lr=slice(1e-5,1e-3))
```

```python
learn.save("SeedRes152_unfreeze_sizeD_lrmed_20")
```

### make lr larger

```python
learn = fv.create_cnn(data, fv.models.resnet152, metrics=fv.error_rate, model_dir = base_dir)
learn.load("SeedRes152_default_16")
learn.unfreeze()
learn.fit_one_cycle(4,max_lr=slice(1e-4,1e-3))
```

```python
learn.save("SeedRes152_unfreeze_sizeD_lrlarger_20")
```
